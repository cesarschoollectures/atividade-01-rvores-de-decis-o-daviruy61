# Questão 1

Utilize o dataset Iris disponível no scikit-learn.
Divida os dados em treino e teste utilizando divisão estratificada.

**Solução**:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Carregar o dataset Iris
iris = load_iris()
X, y = iris.data, iris.target

# Divisão estratificada
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Tamanho do treino: {X_train.shape}")
print(f"Tamanho do teste:  {X_test.shape}")
```

# Questão 2

Treine um modelo utilizando `DecisionTreeClassifier`.

Depois calcule:

- acurácia no treino
- acurácia no teste

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Treinar o modelo
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Acurácia no treino e no teste
acc_train = accuracy_score(y_train, clf.predict(X_train))
acc_test  = accuracy_score(y_test,  clf.predict(X_test))

print(f"Acurácia no treino: {acc_train:.4f}")
print(f"Acurácia no teste:  {acc_test:.4f}")
```

# Questão 3

Utilize `plot_tree()` para visualizar a árvore treinada.

Responda:

**Solução**:

# Questão 4

Treine dez árvores com:

- max_depth = 1
- max_depth = 2
- max_depth = 3
...
- max_depth = 9
- max_depth = None

Registre em uma tabela para cada árvore:

- acurácia no treino
- acurácia no teste
- profundidade da árvore
- número de folhas

**Solução**:

```python
import pandas as pd

resultados = []

profundidades = list(range(1, 10)) + [None]  # 1 a 9 e None

for depth in profundidades:
    modelo = DecisionTreeClassifier(max_depth=depth, random_state=42)
    modelo.fit(X_train, y_train)
    
    acc_treino = accuracy_score(y_train, modelo.predict(X_train))
    acc_teste  = accuracy_score(y_test,  modelo.predict(X_test))
    profundidade = modelo.get_depth()
    n_folhas    = modelo.get_n_leaves()
    
    resultados.append({
        "max_depth": str(depth),
        "Acurácia Treino": round(acc_treino, 4),
        "Acurácia Teste":  round(acc_teste, 4),
        "Profundidade Real": profundidade,
        "Nº de Folhas": n_folhas
    })

df = pd.DataFrame(resultados)
print(df.to_string(index=False))
```

# Questão 5

Treine dois modelos:

- criterion = "gini"
- criterion = "entropy"

Compare:

- profundidade da árvore
- acurácia

**Solução**:

```python
# Treinar dois modelos com critérios diferentes
clf_gini = DecisionTreeClassifier(criterion="gini", random_state=42)
clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)

clf_gini.fit(X_train, y_train)
clf_entropy.fit(X_train, y_train)

# Comparação
print("=" * 45)
print(f"{'Métrica':<25} {'Gini':>8} {'Entropy':>10}")
print("=" * 45)
print(f"{'Acurácia Treino':<25} {accuracy_score(y_train, clf_gini.predict(X_train)):>8.4f} {accuracy_score(y_train, clf_entropy.predict(X_train)):>10.4f}")
print(f"{'Acurácia Teste':<25} {accuracy_score(y_test, clf_gini.predict(X_test)):>8.4f} {accuracy_score(y_test, clf_entropy.predict(X_test)):>10.4f}")
print(f"{'Profundidade da Árvore':<25} {clf_gini.get_depth():>8} {clf_entropy.get_depth():>10}")
print(f"{'Nº de Folhas':<25} {clf_gini.get_n_leaves():>8} {clf_entropy.get_n_leaves():>10}")
print("=" * 45)
```

# Questão 6

Escolha um hiperparâmetro e investigue seu impacto.

Sugestões:

- max_depth
- min_samples_split
- min_samples_leaf
- criterion

Mostre resultados e interprete.
- melhor modelo encontrado
- acurácia
- parâmetros

**Solução**:

```python
import matplotlib.pyplot as plt

parametro = "min_samples_leaf"
valores = [1, 2, 3, 5, 7, 10, 15, 20]

acc_treino_lista = []
acc_teste_lista  = []
profundidades    = []

for v in valores:
    m = DecisionTreeClassifier(min_samples_leaf=v, random_state=42)
    m.fit(X_train, y_train)
    acc_treino_lista.append(accuracy_score(y_train, m.predict(X_train)))
    acc_teste_lista.append(accuracy_score(y_test,  m.predict(X_test)))
    profundidades.append(m.get_depth())

# Tabela de resultados
df_q6 = pd.DataFrame({
    "min_samples_leaf": valores,
    "Acurácia Treino":  [round(a, 4) for a in acc_treino_lista],
    "Acurácia Teste":   [round(a, 4) for a in acc_teste_lista],
    "Profundidade":     profundidades
})
print(df_q6.to_string(index=False))

# Melhor modelo
melhor_idx = df_q6["Acurácia Teste"].idxmax()
melhor = df_q6.iloc[melhor_idx]
print(f"\nMelhor modelo → min_samples_leaf={melhor['min_samples_leaf']}, "
      f"Acurácia Teste={melhor['Acurácia Teste']}, Profundidade={melhor['Profundidade']}")

# Gráfico
plt.figure(figsize=(8, 4))
plt.plot(valores, acc_treino_lista, marker="o", label="Treino")
plt.plot(valores, acc_teste_lista,  marker="s", label="Teste")
plt.xlabel("min_samples_leaf")
plt.ylabel("Acurácia")
plt.title("Impacto do min_samples_leaf na Acurácia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
